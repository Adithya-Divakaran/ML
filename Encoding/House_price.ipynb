```python
import pandas as pd
import category_encoders as ce
```

```python
df = pd.read_csv("/content/drive/MyDrive/Encoding/House_price.csv")
df.head()
```

```python
print(df[['Neighborhood', 'Exterior1st']].head(15))
```

```python
# Binary Encoding
binary_encoder = ce.BinaryEncoder(
    cols=['Neighborhood', 'Exterior1st']
)
df_encoded = binary_encoder.fit_transform(df[['Neighborhood', 'Exterior1st']])
```
