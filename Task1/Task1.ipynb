Question 3

```python
print("Columns:", df.columns.tolist())
```

```python
# Compute required statistics (these ignore NaNs by default)
apple_mean = df['apple(1kg)'].mean()
banana_mean = df['banana(1 dozen)'].mean()
grapes_median = df['grapes(1kg)'].median()
mango_median = df['mango(1kg)'].median()
```

```python
# Create the fill mapping
fill_map = {
    'apple(1kg)': apple_mean,
    'banana(1 dozen)': banana_mean,
    'grapes(1kg)': grapes_median,
    'mango(1kg)': mango_median,
    'Water Melons(1)': 'Not Available'
}
```

```python
# Apply column-wise fill
new_df = df.fillna(fill_map)
```

```python
# Show result
new_df.head(10)
```

Question 4

```python
# Fill null values with the non-null value before it (forward fill)
new_df = df.fillna(method='ffill')
new_df.head(10)
```

Question 5

```python
# Keep rows that have at least 4 non-null values: thresh=4
# (thresh = number of non-NA required to keep the row)
new_df = df.dropna(thresh=4)
new_df
```

Question 6

```python
# Remove rows that contain any null value
new_df = df.dropna()
new_df
```

```python
# Save to 'final_data.csv' with index=False
new_df.to_csv("final_data.csv", index=False)
print("Saved final_data.csv (index=False).")
```
